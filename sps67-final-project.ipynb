When analyzing the Allegheny County Assets data set, I developed a few metrics to interpret this data. Basically, the assets can be categorized as Charity, for the Elderly, Recreational, Educational, Healthcare, Buisness, and Public works. Each of these categories is given a base point value with some flexibility to adjust case by case. This point value determines its overall weight upon finding the best neighborhood. The neighborhood with the highest overall value between all of its assets is deemed most desireable. 

### Here is the point value breakdown for each category:
* Charity: -5
* For the elderly: -2
* Recreational: +3
* Educational: +3
* Healthcare: +2
* Buisness: +1
* Public works: +1

### Expected Neighborhood
Based on how the asset metrics are defined, the overall best neighborhood will have many assets compared to the other neighborhoods.
```python
import pandas as pd
import numpy as np

# Read in data from database
assets = pd.read_csv("https://data.wprdc.org/datastore/dump/5c7825d2-6814-40c7-aefe-3d0f3d6f22e7", low_memory = False)

# Get only data from Pittsburgh
query_mask = assets["city"] == "Pittsburgh"
pittsburgh = assets[query_mask]

# Drop duplicate assets
pittsburgh = pittsburgh.drop_duplicates(subset=["name"], keep='last')

# Dictionary to hold scores for each zipcode
scores = {}

# Function to calculate score for asset_type
def calculate_score(type):
    if type == "affordable_housing" or type == "community_nonprofit_orgs" or type == "food_banks":
        return -5
    elif type == "nursing_homes" or type == "va_facilities" or type == "senior_centers" or type == "veterans_social_orgs":
        return -2
    elif type == "parks_and_facilities" or type == "schools" or type == "park_and_rides" or type == "bars" or type == "libraries" or type == "rec_centers":
        return 3
    elif type == "dentists" or type == "doctors_offices" or type == "health_centers" or type == "pharmacies" or type == "hotels" or type == "achd_clinics":
        return 2
    else:
        return 5

# Calculate scores for assets
for asset in pittsburgh.index:
    type = pittsburgh['asset_type'][asset]
    if isinstance(pittsburgh['zip_code'][asset], str):
        zip = pittsburgh['zip_code'][asset][:]
        score = calculate_score(type)
        if zip in scores:
            scores[zip] += score
        else:
            scores[zip] = score

# Find zip with highest score
max = 0
max_key = 0
for key in scores:
    if scores[key] > max and len(key) > 5:
        max_key = key
        max = scores[key]
print(max)
print(max_key)
```
